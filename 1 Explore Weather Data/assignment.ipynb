```python
import pandas as pd
```

```python
bd = pd.read_csv('berlin_data.csv')
gd = pd.read_csv('global_data.csv')
```

```python
bd.head(20)
```

```python
bd.shape
```

```python
gd.head()
```

```python
gd.shape
```

```python
bd['ma']=bd.rolling(window=10)['avg_temp'].mean()
gd['ma']=gd.rolling(window=10)['avg_temp'].mean()
```

```python
bd.head(20)
```

```python
gd.head(20)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig = plt.figure()

ax = fig.add_subplot(111)
ax.plot(gd['year'], gd['ma'], label='Global')
ax.plot(bd['year'], bd['ma'], label='Berlin')
ax.set(title='Weather data',
      ylabel='Temperature in C',
      xlabel='Year')
ax.legend(loc='best')

plt.savefig('temps.svg')
plt.show()
```

```python
ld = pd.read_csv('la_data.csv')
sd = pd.read_csv('shanghai_data.csv')
fd = pd.read_csv('sanfran_data.csv')
ld['ma']=ld.rolling(window=10)['avg_temp'].mean()
sd['ma']=sd.rolling(window=10)['avg_temp'].mean()
fd['ma']=fd.rolling(window=10)['avg_temp'].mean()
```

```python
fig2 = plt.figure()

ax2 = fig2.add_subplot(111)
ax2.plot(gd['year'], gd['ma'], label='Global')
ax2.plot(bd['year'], bd['ma'], label='Berlin')
#ax2.plot(ld['year'], ld['ma'], label='Los Angeles')
ax2.plot(sd['year'], sd['ma'], label='Shanghai')
ax2.plot(fd['year'], fd['ma'], label='San Francisco')
ax2.set(title='Weather data',
      ylabel='Temperature in C',
      xlabel='Year')
ax2.legend(loc='best')

plt.savefig('temps2.svg')
plt.show()
```
